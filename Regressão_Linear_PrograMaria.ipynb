```python
pd.get_dummies(dados, columns =["GENERO", "REGIAO ONDE MORA", "NOVO_NIVEL", "SETOR"], drop_first=True)
# Cria variáveis dummies (também chamadas de variáveis indicadoras) para as colunas categóricas:
# "GENERO", "REGIAO ONDE MORA", "NOVO_NIVEL" e "SETOR".
# Isso transforma cada categoria dessas colunas em uma nova coluna binária (0 ou 1),
# facilitando o uso desses dados em modelos estatísticos ou de machine learning.
# O parâmetro drop_first=True remove a primeira categoria de cada coluna
# para evitar a multicolinearidade (problema de redundância) nos modelos.
# O resultado é um novo DataFrame com as colunas originais convertidas em colunas numéricas.
```

```python
dados = pd.get_dummies(dados, columns =["GENERO", "REGIAO ONDE MORA", "NOVO_NIVEL", "SETOR"], drop_first=True)
```

```python
X = dados.drop("SALARIO", axis=1)
Y = dados["SALARIO"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
# Essa linha divide os dados em 4 conjuntos:
# - X_train: dados de entrada (features) para treino do modelo.
# - X_test: dados de entrada para teste do modelo (avaliar desempenho).
# - Y_train: rótulos (valores alvo) correspondentes ao X_train.
# - Y_test: rótulos correspondentes ao X_test.

# A função train_test_split faz essa separação:
# - X e Y são, respectivamente, as variáveis independentes (features) e a dependente (target).
# - test_size=0.2 indica que 20% dos dados serão usados para teste e 80% para treino.
# - random_state=42 serve para garantir que a separação seja sempre a mesma quando o código for executado — isso é importante para reprodutibilidade.
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train_scaled, Y_train)
```

```python
model.predict(X_test_scaled)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
mse = mean_squared_error(Y_test, y_pred)
mse
```

```python
mae = mean_absolute_error(Y_test, y_pred)
mae
```

```python
r2 = r2_score(Y_test, y_pred)
r2
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(Y_test, y_pred, alpha=0.5)
plt.xlabel("Valor Real")
plt.ylabel("Valor Predito")
plt.title("Comparação entre Valores Reais e Previstos")
plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)], linestyle='--', color='blue')
plt.show()
```
