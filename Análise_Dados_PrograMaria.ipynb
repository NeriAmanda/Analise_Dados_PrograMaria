### Correlação

- A correlação nos diz se e como duas variáveis estão relacionadas. Existem dois tipos principais de correlação: positiva e negativa.
- A correlação positiva ocorre quando duas variáveis aumentam juntas.
- Por exemplo, quanto mais um aluno estuda, maior é seu desempenho acadêmico.Já a correlação negativa acontecia quando uma variável aumenta enquanto a outra diminui.

```python
# Vamos calcular a correlação IDADE/SALARIO
correlacao_continua = dados['IDADE'].corr(dados['SALARIO'])
correlacao_continua
# O valor final ficou em 0,2, ou seja mais proximo do 0 do que do 1 mostrando pouca relação entre IDADE e SALARIO
```

```python
from scipy.stats import chi2_contingency
```

```python
def cramer_coeficiente(coluna1, coluna2): # Gera uma tabela de contingência (frequência cruzada) entre as duas variáveis
  tabela_cruzada = np.array (pd.crosstab(coluna1, coluna2))
  chi2 = chi2_contingency(tabela_cruzada)[0] # Calcula o valor do qui-quadrado a partir da tabela de contingência
  soma = np.sum(tabela_cruzada) # Soma total de observações da tabela
  mini = min(tabela_cruzada.shape)-1 # Determina o menor número de categorias entre as duas variáveis, subtraindo 1
  cramer = np.sqrt(chi2/(soma*mini)) # Calcula o coeficiente de Cramér, que mede a associação entre variáveis categóricas
  return cramer
```

```python
# Com esse resultados (0,04) podemos dizer que as colunas de COR/RACA/ETNIA e NIVEL DE ENSINO não tem quase nenhuma relação
cramer_coeficiente(dados['COR/RACA/ETNIA'], dados['NIVEL DE ENSINO'])
```

Gráfico de Dispersão:

Um gráfico de dispersão mostra a relação entre duas variáveis numéricas, como idade e salário. Cada ponto no gráfico representa uma pessoa, indicando sua idade no eixo X e seu salário no eixo Y.

- Quando os pontos estão **mais próximos e alinhados**, isso significa que existe uma relação forte entre as variáveis. Por exemplo, o salário tende a aumentar conforme a idade aumenta.
- Quando os pontos estão **mais dispersos e espalhados**, isso indica que a relação entre idade e salário é fraca ou pouco clara, ou seja, a idade não explica muito o valor do salário.

O gráfico ajuda a visualizar se existe um padrão ou tendência entre as variáveis, facilitando a análise dos dados.


```python
plt.figure(figsize=(20, 6))  # Define o tamanho da figura para melhor visualização
# Cria um gráfico de dispersão (scatter plot) onde:
# O eixo X representa a idade das pessoas (dados['IDADE'])
# O eixo Y representa o salário correspondente (dados['SALARIO'])
# alpha=0.5 deixa os pontos semi-transparentes, facilitando a visualização em áreas com muitos dados sobrepostos
plt.scatter(dados['IDADE'], dados['SALARIO'], alpha=0.5)
plt.title('Dispersão de Salário Médio por Idade')
plt.xlabel('Idade')
plt.ylabel('Salário Médio')
plt.grid(True)
plt.show()
```

```python
# Define os limites das faixas etárias que queremos criar
bins = [0, 25, 35, 45, 55, 100]

# Define os nomes (rótulos) correspondentes para cada faixa etária
labels = ['0-25', '26-35', '36-45', '46-55', '56+']

# Cria uma nova coluna 'Faixa_Idade' no DataFrame 'dados' que categoriza cada idade dentro das faixas definidas
# O parâmetro right=False indica que o limite superior da faixa não está incluído (exemplo: 25 não entra na faixa 0-25)
dados['Faixa_Idade'] = pd.cut(dados['IDADE'], bins=bins, labels=labels, right=False)

# Conta quantas pessoas existem em cada faixa etária e transforma essa contagem em um novo DataFrame
faixa_contagem = dados['Faixa_Idade'].value_counts().reset_index()

# Renomeia as colunas do novo DataFrame para nomes mais claros
# 'Faixa_Idade' para as categorias e 'Quantidade' para o número de pessoas em cada faixa
faixa_contagem.columns = ['Faixa_Idade', 'Quantidade']

# Cria o gráfico de pizza usando Plotly Express, mostrando a proporção de pessoas em cada faixa etária
# names: categorias (faixas etárias)
# values: quantidade de pessoas por categoria
# title: título do gráfico
fig = px.pie(faixa_contagem, names='Faixa_Idade', values='Quantidade', title='Proporção de Pessoas por Faixa Etária')

# Exibe o gráfico gerado
fig.show()
```
